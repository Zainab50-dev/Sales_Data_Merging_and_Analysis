```python
# Splitting, Cleaning, Merging and Analyzing Multiple Sales Files
```

```python
import pandas as pd
```

```python
# Loading Dataset
```

```python
df = pd.read_csv('Superstore.csv', encoding = "ISO-8859-1")
```

```python
df.head(30)
```

```python
# Splitting into 4 files
```

```python
num_splits = 4
rows_per_split = len(df) // num_splits
for i in range(num_splits):
    start = i * rows_per_split
    end = (i + 1) * rows_per_split
    split_df = df.iloc[start : end]
    split_df.to_excel(f"part{i+1}.xlsx", index = False)
    print(f"Creating part{i+1} with rows{start} to {end-1}")
```

```python
# Cleaning Files
```

```python
files = ["part1.xlsx","part2.xlsx","part3.xlsx","part4.xlsx"]
for file in files:
    Df = pd.read_excel(file)
    Df = Df.drop(columns=['Row ID'])
    Df['Order Date'] = pd.to_datetime(df['Order Date'])
    Df['Ship Date'] = pd.to_datetime(df['Ship Date'])
    cleaned_file_name = file.replace(".xlsx", "clean.xlsx")
    Df.to_excel(cleaned_file_name, index = False)
```

```python
import glob
```

```python
all_files = glob.glob("part*clean.xlsx")
```

```python
dfs = []
```

```python
for file in all_files:
    dff = pd.read_excel(file)
    dfs.append(dff)
```

```python
final_df = pd.concat(dfs,ignore_index = True)
```

```python
final_df.to_excel("merged_output.xlsx", index = False)
```

```python
# Cleaned Merged Dataset
```

```python
final_df
```

```python
## Insights
```

```python
region_sales = final_df.groupby('Region')['Sales'].sum().reset_index().sort_values(by = 'Sales',ascending = False)
region_sales
```

```python
category_sales = final_df.groupby('Category')['Sales'].sum().reset_index().sort_values(by = 'Sales',ascending = False)
category_sales
```

```python
product_wise_profit = final_df.groupby('Product Name')['Profit'].sum().reset_index().sort_values(by = 'Profit', ascending = False)
product_wise_profit
```

```python
category_wise_profit = final_df.groupby('Category')['Profit'].sum().reset_index().sort_values(by = 'Profit',ascending = False)
category_wise_profit
```

```python
top_customers = final_df.groupby('Customer Name')['Sales'].sum().reset_index().sort_values(by = 'Sales',ascending = False)
top_customers.head(10)
```

```python
top_customers_byprofit = final_df.groupby('Customer Name')['Profit'].sum().reset_index().sort_values(by = 'Profit',ascending = False)
top_customers_byprofit
```

```python
# Creating Month/Year columns
```

```python
final_df['Month'] = final_df['Order Date'].dt.month
final_df['Year'] = final_df['Order Date'].dt.year
```

```python
monthly_sales = final_df.groupby(['Year','Month'])['Sales'].sum().reset_index()
monthly_sales
```

```python
yearly_sales = final_df.groupby('Year')['Sales'].sum().reset_index()
yearly_sales
```

```python
final_df['Profit_Margin'] = ((final_df['Profit']/ final_df['Sales']) * 100)
final_df['Profit_Margin']
```

```python
# Exporting Cleaned Dataset To Excel
```

```python
final_df.to_excel("project2.xlsx",sheet_name = "Cleaned_data",index = False)
```

```python
# Exporting insghts to excel in separate sheets
```

```python
with pd.ExcelWriter("project2_insightz.xlsx",engine = "openpyxl") as writer:
    final_df.to_excel(writer,sheet_name = "Cleaned_Data", index = False)
    region_sales.to_excel(writer,sheet_name = "Region_Sales", index = False)
    category_sales.to_excel(writer,sheet_name = "Category_Sales", index = False)
    product_wise_profit.to_excel(writer,sheet_name = "product_wise_profit", index = False)
    category_wise_profit.to_excel(writer,sheet_name = "category_wise_profit", index = False)
    top_customers.to_excel(writer,sheet_name = "top_customers", index = False)
    top_customers_byprofit.to_excel(writer,sheet_name = "top_customers_byprofit", index = False)
    monthly_sales.to_excel(writer,sheet_name = "monthly_sales", index = False)
    yearly_sales.to_excel(writer,sheet_name = "yearly_sales", index = False)
```
